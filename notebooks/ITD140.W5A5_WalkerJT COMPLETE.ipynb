## **Task 1: Number of students in different major groups** - 10 points
- Objective: Generate a bar chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path (use raw string r'' to avoid escape character issues)
file_path = r'C:\CollegeData.csv'

# Read the CSV file
college_data = pd.read_csv(file_path)

# Mapping major groups to their long labels
major_group_mapping = {
    0: 'Undeclared',
    1: 'Engineering',
    2: 'Science and Math',
    3: 'Social Sciences and Humanities',
    101: 'Unknown'
}

# Ensure 'MAJ_GRP' is treated as an integer before mapping
college_data['MAJ_GRP'] = college_data['MAJ_GRP'].astype(int)
college_data['MajorGroup'] = college_data['MAJ_GRP'].map(major_group_mapping)

# Aggregating the number of students by major group
major_group_counts = college_data['MajorGroup'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(major_group_counts.index, major_group_counts.values, color='blue')

# Adding chart labels and title
plt.xlabel('Major Group')
plt.ylabel('Number of Students')
plt.title('Number of Students in Different Major Groups')
plt.xticks(rotation=45)  # Rotate labels for readability
plt.legend(['Students'])

# Display the chart
plt.show()
```

## **Task 2: SAT Math scores over time.** - 10 Points
- Objective: Generate a line chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
# Insert your code below this text and execute the code.

# Line chart of SAT Math scores over time
avg_sat_math_by_year = college_data.groupby('GRAD_YR')['SAT_M'].mean()

plt.plot(avg_sat_math_by_year.index, avg_sat_math_by_year.values, marker='o', linestyle='-', color='green')
plt.xlabel('Graduation Year')
plt.ylabel('Average SAT Math Score')
plt.title('Trend of Average SAT Math Scores by Graduation Year')
plt.show()
```

## **Task 3: Gender distribution** - 10 Points
- Objective: Generate a pie chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Counting the number of students by gender category
gender_counts = college_data['GENDER_C'].value_counts()

# Define labels and colors
labels = ['Male', 'Female']
colors = ['lightblue', 'lightpink']

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts.values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Adding a title
plt.title('Proportion of Students by Gender Category')

# Adding a legend
plt.legend(labels, title="Gender", loc="best")

# Display the chart
plt.show()
```

## **Task 7: Cumulative number of students in different major groups over time** - 10 Points
- Objective: Generate an area chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
# Insert your code below this text and execute the code.

# Map MAJ_GRP numeric values to text labels
maj_grp_mapping = {
    0: 'Undeclared',
    1: 'Engineering',
    2: 'Science and Math',
    3: 'Social Sciences and Humanities',
    101: 'Unknown'
}
college_data['MAJ_GRP'] = college_data['MAJ_GRP'].map(maj_grp_mapping)

# Group by 'GRAD_YR' and 'MAJ_GRP' and count the number of students in each group
grouped_college_data = college_data.groupby(['GRAD_YR', 'MAJ_GRP']).size().unstack(fill_value=0)

# Calculate the cumulative sum over time
cumulative_college_data = grouped_college_data.cumsum()

# Generate the area chart
cumulative_college_data.plot(kind='area', stacked=True, figsize=(12, 8))

# Add labels and title
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.title('Cumulative Number of Students in Different Major Groups Over Time')

# Add a legend
plt.legend(title='Major Groups')

# Show the plot
plt.show()
```

## **Task 8: Ethnicity distribution within different major groups** - 10 Points
- Objective: Generate a stacked bar chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
# Insert your code below this text and execute the code.

import pandas as pd
import matplotlib.pyplot as plt

# Ensure categorical columns are treated correctly
college_data['MAJ_GRP'] = college_data['MAJ_GRP'].astype(str)
college_data['ETHNICITY'] = college_data['ETHNICITY'].astype(str)

# Group by major group and ethnicity, then count the number of students
ethnicity_counts = college_data.groupby(['MAJ_GRP', 'ETHNICITY']).size().unstack(fill_value=0)

# Create a stacked bar chart
ethnicity_counts.plot(kind='bar', stacked=True, colormap='viridis')

# Add labels and title
plt.xlabel('Major Group')
plt.ylabel('Number of Students')
plt.title('Distribution of Ethnicity Within Major Groups')

# Display the chart
plt.show()
```

## **Task 9: Correlation matrix between numerical variables** - 10 Points
- Objective: Generate a heat map that depicts the given task. 
- Instruction:
  - Use `Seaborn` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

**IMPORTANT:** To obtain credit for this question, the numbers in the heatmap must be readable. HINT: Adjust  `linewidths`, `fmt` and `annot_kws` to manipulate text, decimals and cell spacing.

```python
# Insert your code below this text and execute the code.

import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = college_data.corr()

# Set larger figure size for better readability
plt.figure(figsize=(12, 10))

# Create a heat map using Seaborn with improved settings
sns.heatmap(correlation_matrix, 
            annot=True,             # Show numbers inside cells
            fmt=".2f",              # Limit decimal places
            cmap="coolwarm",        # Use color scheme for better contrast
            linewidths=1.0,         # Add space between cells
            annot_kws={"size": 8},  # Reduce annotation font size
            square=True             # Keep cells square-shaped
)

# Rotate labels for better readability
plt.xticks(rotation=45, ha="right")  # Rotate X labels
plt.yticks(rotation=0)  # Keep Y labels horizontal

# Add title
plt.title('Heat Map of Correlations Between Numeric Variables')

# Display the chart
plt.show()
```

## **Task 10: Compare mean SAT Math scores between different feeder schools** - 10 Points
- Objective: Generate a grouped bar chart that depicts the given task. 
- Instruction:
  - Use `matplotlib` to generate the required chart.
  - Use appropriate chart labels and legends.
  - Ensure data preprocessing / cleaning.

### <span style="color: blue">Insert your code below this text and execute the code - 10 points:</span>

```python
# Insert your code below this text and execute the code.

import pandas as pd
import matplotlib.pyplot as plt

# Ensure categorical columns are treated correctly
college_data['FEEDER_C'] = college_data['FEEDER_C'].astype(str)

# Group by feeder school and calculate mean SAT Math scores
mean_sat_by_feeder = college_data.groupby('FEEDER_C', as_index=False)['SAT_M'].mean()

# Create a grouped bar chart
plt.bar(mean_sat_by_feeder['FEEDER_C'], mean_sat_by_feeder['SAT_M'], color='lightcoral')

# Add labels and title
plt.xlabel('Feeder School')
plt.ylabel('Mean SAT Math Score')
plt.title('Mean SAT Math Scores by Feeder School')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```
